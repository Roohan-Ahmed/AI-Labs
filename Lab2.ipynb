```python
#Task 1
import random

class Environment:
    def __init__(self):
        self.components = {chr(65+i):random.choice(['safe', 'vulnerable']) for i in range(9)}

    def display_state(self):
        print("System State:")
        for component, status in self.components.items():
            print(f"Component {component}: {status}")
        print()

    def get_percept(self):
        return self.state

class Scan:
    def __init__(self, Environment):
        self.environment = Environment


    def scanner(self):
        for component in self.environment.components:
            if self.environment.components[component] == 'vulnerable':
                print(f"Warning: Component {component} is vulnerable!")
            else:
              print(f"Success: Component {component} is safe.")
        print()

    def patcher(self):
        for component in self.environment.components:
            if self.environment.components[component] == 'vulnerable':
                self.environment.components[component] == 'safe'
                print(f"Warning: Component {component} is now safe!")
            else:
                print(f"Success: Component {component} is safe.")
        print()

env = Environment()
env.display_state()
agent = Scan(env)
agent.scanner()
agent.patcher()
```

```python
#Task 2
import random

class Environment:
    def __init__(self, name, num_servers=5):
        self.name = name
        self.Servers = {f"Server-{i+1}": random.choice(["Underloaded", "Balanced", "Overloaded"]) for i in range(num_servers)}

class LoadBalancerAgent:
    def __init__(self):
        pass

    def scan(self, servers):
        print("Scanning servers:")
        for server, load in servers.items():
            if load == "Underloaded" or load == "Overloaded":
                print(f"Load not balanced in {server}: {load}")
            else:
                print(f"{server} is balanced")
        print()

    def act(self, servers):
        print("Balancing servers ::")
        for server in servers:
            if servers[server] in ["Underloaded", "Overloaded"]:
                servers[server] = "Balanced"
        print("Balancing complete!\n")

def agent_call(agent, environment):
    agent.scan(environment.Servers)
    agent.act(environment.Servers)
    agent.scan(environment.Servers)

environment = Environment("Data Center")
Agent = LoadBalancerAgent()

agent_call(Agent, environment)
```

```python
#Task 3
import random

class Environment:
    def __init__(self, name, num_tasks=5):
        self.name = name
        self.BackupTasks = {f"Task-{i+1}": random.choice(["Completed", "Failed"]) for i in range(num_tasks)}

class BackupManagementAgent:
    def __init__(self):
        pass

    def scan(self, tasks):
        print("Scanning backup tasks:")
        for task, status in tasks.items():
            if status == "Failed":
                print(f"Backup {task} has failed.")
            else:
                print(f"{task} is completed.")
        print()

    def retry(self, tasks):
        print("Retrying failed backups ::")
        for task in tasks:
            if tasks[task] == "Failed":
                tasks[task] = "Completed"
        print("Retrying completed.\n")

def agent_call(agent, network):
    agent.scan(network.BackupTasks)
    agent.retry(network.BackupTasks)
    agent.scan(network.BackupTasks)

network = Environment("Server Network")
Agent = BackupManagementAgent()

agent_call(Agent, network)
```

```python
#Task 4
import random

class Environment:
    def __init__(self):
        self.components = {comp: random.choice(["Safe", "Low Risk Vulnerable", "High Risk Vulnerable"]) for comp in "ABCDEFGHI"}

class UtilityBasedSecurityAgent:
    def __init__(self):
        pass

    def scan(self, system):
        print("Scanning security system:")
        for component, status in system.components.items():
            if status == "Safe":
                print(f"{component} is Safe.")
            else:
                print(f"WARNING: {component} has a {status}.")
        print()

    def patch(self, system):
        print("Patching vulnerabilities :: ")
        for i in system.components:
            if system.components[i] == "Low Risk Vulnerable":
                system.components[i] = "Safe"
                print(f"Patched {i} (Low Risk Vulnerability).")
            elif system.components[i] == "High Risk Vulnerable":
                print(f"{i} is high risk vulnerable. Premium service required.")
        print("Patching process completed.\n")

def agent_call(agent, system):
    print("Initial System State:")
    agent.scan(system)
    agent.patch(system)
    print("Final System State:")
    agent.scan(system)

security_system = Environment()
Agent = UtilityBasedSecurityAgent()

agent_call(Agent, security_system)
```

```python
#Task 5
import random

class HospitalEnvironment:
    def __init__(self):
        self.rooms = {f"Room-{i+1}": random.choice(["Needs Medicine", "No Delivery Required"]) for i in range(5)}
        self.medicine_storage = "Available"
        self.nurse_station = "Staff Available"

    def display_rooms_status(self):
        print("\nCurrent Room Status:")
        for room, status in self.rooms.items():
            print(f"{room}: {status}")
        print()

class HospitalDeliveryRobot:
    def __init__(self):
        self.location = "Start"
        self.medicine = None

    def move_to(self, destination):
        print(f"Moving to {destination}...")
        self.location = destination

    def pick_up_medicine(self, environment):
        if environment.medicine_storage == "Available":
            self.medicine = "Medicine"
            print("Picked up medicine from storage.")
        else:
            print("Medicine storage is empty!")

    def scan_patient_id(self, room):
        print(f"Scanning patient ID in {room}.")
        return True

    def deliver_medicine(self, environment, room):
        if self.medicine and environment.rooms[room] == "Needs Medicine":
            if self.scan_patient_id(room):
                print(f"Delivered medicine to {room}.")
                environment.rooms[room] = "No Delivery Required"
                self.medicine = None
        else:
            print(f"No delivery needed for {room}.")

def agent_call(agent, environment):
    print("Starting hospital delivery process :\n")
    agent.move_to("Medicine Storage")
    agent.pick_up_medicine(environment)

    for room in environment.rooms:
        agent.move_to(room)
        agent.deliver_medicine(environment, room)
    print("Delivery process completed.")

hospital_env = HospitalEnvironment()
robot_agent = HospitalDeliveryRobot()

agent_call(robot_agent, hospital_env)
```

```python
class FirefightingRobot:
    def __init__(self):
        self.grid = {
            'a': "Safe", 'b': "Safe", 'c': "Fire",
            'd': "Safe", 'e': "Fire", 'f': "Safe",
            'g': "Safe", 'h': "Safe", 'j': "Fire"
        }
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
        self.position = 'a'

    def display_environment(self):
        print("\nCurrent Environment Status:")
        for i, room in enumerate(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']):
            symbol = "游댠" if self.grid[room] == "Fire" else "游릴"
            print(symbol, end=" " if (i + 1) % 3 != 0 else "\n")
        print()

class Agent:
    def __init__(self, robot):
        self.robot = robot

    def act(self):
        for i in self.robot.path:
            self.robot.position = i
            print(f"Robot moved to {i}.")
            if self.robot.grid[i] == "Fire":
                print(f"Fire detected in {i}!!")
                self.robot.grid[i] = "Safe"
                print(f"Fire in {i} extinguished.")
            self.robot.display_environment()
        print("All rooms checked. Final environment status:")
        self.robot.display_environment()

robot = FirefightingRobot()
agent = Agent(robot)
robot.display_environment()
agent.act()
```
